```python
# MNIST Digit Classification - Google Colab Implementation
# Author: sntsemilio
# Date: 2025-06-22

# ============================================================================
# SECTION 1: Setup and Imports
# ============================================================================

# Install required packages (if needed)
!pip install streamlit tensorflow matplotlib seaborn scikit-learn

# Import libraries
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd
from datetime import datetime

print("TensorFlow version:", tf.__version__)
print("GPU Available:", tf.config.list_physical_devices('GPU'))

# ============================================================================
# SECTION 2: Data Loading and Preprocessing
# ============================================================================

def load_and_preprocess_mnist():
    """Load and preprocess MNIST dataset"""
    print("Loading MNIST dataset...")

    # Load MNIST dataset
    (x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

    # Normalize pixel values to [0, 1]
    x_train = x_train.astype('float32') / 255.0
    x_test = x_test.astype('float32') / 255.0

    # Reshape for CNN (add channel dimension)
    x_train = x_train.reshape(-1, 28, 28, 1)
    x_test = x_test.reshape(-1, 28, 28, 1)

    # Create validation split
    from sklearn.model_selection import train_test_split
    x_train, x_val, y_train, y_val = train_test_split(
        x_train, y_train, test_size=0.1, random_state=42, stratify=y_train
    )

    print(f"Training samples: {len(x_train)}")
    print(f"Validation samples: {len(x_val)}")
    print(f"Test samples: {len(x_test)}")
    print(f"Image shape: {x_train[0].shape}")

    return (x_train, y_train), (x_val, y_val), (x_test, y_test)

# Load the data
(x_train, y_train), (x_val, y_val), (x_test, y_test) = load_and_preprocess_mnist()

# ============================================================================
# SECTION 3: Data Visualization
# ============================================================================

def visualize_dataset():
    """Visualize dataset samples and statistics"""

    # Plot sample images
    plt.figure(figsize=(15, 6))

    # Random samples
    plt.subplot(1, 3, 1)
    fig, axes = plt.subplots(4, 5, figsize=(10, 8))
    fig.suptitle('Random MNIST Samples', fontsize=16)

    indices = np.random.choice(len(x_train), 20, replace=False)
    for i, idx in enumerate(indices):
        row, col = i // 5, i % 5
        axes[row, col].imshow(x_train[idx].squeeze(), cmap='gray')
        axes[row, col].set_title(f'Label: {y_train[idx]}')
        axes[row, col].axis('off')

    plt.tight_layout()
    plt.show()

    # Class distribution
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    class_counts = np.bincount(y_train)
    plt.bar(range(10), class_counts, color='skyblue', alpha=0.7)
    plt.title('Training Set Class Distribution')
    plt.xlabel('Digit')
    plt.ylabel('Count')
    for i, count in enumerate(class_counts):
        plt.text(i, count + 50, str(count), ha='center', va='bottom')

    plt.subplot(1, 2, 2)
    test_counts = np.bincount(y_test)
    plt.bar(range(10), test_counts, color='lightcoral', alpha=0.7)
    plt.title('Test Set Class Distribution')
    plt.xlabel('Digit')
    plt.ylabel('Count')
    for i, count in enumerate(test_counts):
        plt.text(i, count + 10, str(count), ha='center', va='bottom')

    plt.tight_layout()
    plt.show()

# Visualize the dataset
visualize_dataset()

# ============================================================================
# SECTION 4: Model Definition
# ============================================================================

def create_mnist_model():
    """Create a CNN model for MNIST classification"""

    model = keras.Sequential([
        # First Convolutional Block
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
        layers.MaxPooling2D((2, 2)),

        # Second Convolutional Block
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),

        # Third Convolutional Block
        layers.Conv2D(64, (3, 3), activation='relu'),

        # Flatten and Dense layers
        layers.Flatten(),
        layers.Dense(64, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(10, activation='softmax')
    ])

    # Compile the model
    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    return model

# Create the model
print("Creating MNIST CNN model...")
model = create_mnist_model()

# Display model architecture
model.summary()

# Visualize model architecture
tf.keras.utils.plot_model(model, show_shapes=True, show_layer_names=True, dpi=150)

# ============================================================================
# SECTION 5: Model Training
# ============================================================================

def train_model(model, x_train, y_train, x_val, y_val, epochs=10):
    """Train the model with callbacks"""

    print(f"Starting training for {epochs} epochs...")

    # Define callbacks
    callbacks = [
        keras.callbacks.EarlyStopping(
            patience=3,
            restore_best_weights=True,
            verbose=1
        ),
        keras.callbacks.ReduceLROnPlateau(
            factor=0.2,
            patience=2,
            min_lr=0.0001,
            verbose=1
        ),
        keras.callbacks.ModelCheckpoint(
            'best_mnist_model.h5',
            save_best_only=True,
            verbose=1
        )
    ]

    # Train the model
    start_time = datetime.now()

    history = model.fit(
        x_train, y_train,
        batch_size=32,
        epochs=epochs,
        validation_data=(x_val, y_val),
        callbacks=callbacks,
        verbose=1
    )

    end_time = datetime.now()
    training_time = end_time - start_time

    print(f"\nTraining completed in: {training_time}")

    return history

# Train the model
history = train_model(model, x_train, y_train, x_val, y_val, epochs=15)

# ============================================================================
# SECTION 6: Training Visualization
# ============================================================================

def plot_training_history(history):
    """Plot training history"""

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # Plot accuracy
    ax1.plot(history.history['accuracy'], label='Training Accuracy', marker='o')
    ax1.plot(history.history['val_accuracy'], label='Validation Accuracy', marker='s')
    ax1.set_title('Model Accuracy')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Accuracy')
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    # Plot loss
    ax2.plot(history.history['loss'], label='Training Loss', marker='o')
    ax2.plot(history.history['val_loss'], label='Validation Loss', marker='s')
    ax2.set_title('Model Loss')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Loss')
    ax2.legend()
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    # Print final metrics
    final_train_acc = history.history['accuracy'][-1]
    final_val_acc = history.history['val_accuracy'][-1]
    final_train_loss = history.history['loss'][-1]
    final_val_loss = history.history['val_loss'][-1]

    print(f"\nFinal Training Accuracy: {final_train_acc:.4f}")
    print(f"Final Validation Accuracy: {final_val_acc:.4f}")
    print(f"Final Training Loss: {final_train_loss:.4f}")
    print(f"Final Validation Loss: {final_val_loss:.4f}")

# Plot training results
plot_training_history(history)

# ============================================================================
# SECTION 7: Model Evaluation
# ============================================================================

def evaluate_model(model, x_test, y_test):
    """Comprehensive model evaluation"""

    print("Evaluating model on test set...")

    # Basic evaluation
    test_loss, test_accuracy = model.evaluate(x_test, y_test, verbose=0)
    print(f"Test Loss: {test_loss:.4f}")
    print(f"Test Accuracy: {test_accuracy:.4f}")

    # Predictions
    y_pred_proba = model.predict(x_test, verbose=0)
    y_pred = np.argmax(y_pred_proba, axis=1)

    # Classification report
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, digits=4))

    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)

    # Plot confusion matrix
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=range(10), yticklabels=range(10))
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    # Per-class accuracy
    class_accuracy = cm.diagonal() / cm.sum(axis=1)

    plt.figure(figsize=(10, 6))
    bars = plt.bar(range(10), class_accuracy, color='lightgreen', alpha=0.7)
    plt.title('Per-Class Accuracy')
    plt.xlabel('Digit')
    plt.ylabel('Accuracy')
    plt.ylim(0, 1)

    # Add accuracy labels on bars
    for i, acc in enumerate(class_accuracy):
        plt.text(i, acc + 0.01, f'{acc:.3f}', ha='center', va='bottom')

    plt.grid(axis='y', alpha=0.3)
    plt.show()

    return {
        'test_accuracy': test_accuracy,
        'test_loss': test_loss,
        'predictions': y_pred,
        'probabilities': y_pred_proba,
        'confusion_matrix': cm,
        'class_accuracy': class_accuracy
    }

# Evaluate the model
results = evaluate_model(model, x_test, y_test)

# ============================================================================
# SECTION 8: Error Analysis
# ============================================================================

def analyze_errors(model, x_test, y_test, results):
    """Analyze misclassified examples"""

    y_pred = results['predictions']

    # Find misclassified examples
    misclassified_mask = y_pred != y_test
    misclassified_indices = np.where(misclassified_mask)[0]

    print(f"Total misclassified examples: {len(misclassified_indices)}")
    print(f"Error rate: {len(misclassified_indices)/len(y_test):.4f}")

    if len(misclassified_indices) > 0:
        # Show some misclassified examples
        num_examples = min(15, len(misclassified_indices))
        selected_indices = np.random.choice(misclassified_indices, num_examples, replace=False)

        fig, axes = plt.subplots(3, 5, figsize=(15, 9))
        fig.suptitle('Misclassified Examples', fontsize=16)

        for i, idx in enumerate(selected_indices):
            row, col = i // 5, i % 5

            # Get prediction probabilities
            proba = results['probabilities'][idx]
            confidence = np.max(proba)

            axes[row, col].imshow(x_test[idx].squeeze(), cmap='gray')
            axes[row, col].set_title(
                f'True: {y_test[idx]}, Pred: {y_pred[idx]}\n'
                f'Conf: {confidence:.2f}',
                fontsize=10
            )
            axes[row, col].axis('off')

        plt.tight_layout()
        plt.show()

# Analyze errors
analyze_errors(model, x_test, y_test, results)

# ============================================================================
# SECTION 9: Interactive Prediction Function
# ============================================================================

def predict_digit(model, image_index=None):
    """Predict a digit from test set or random selection"""

    if image_index is None:
        image_index = np.random.randint(0, len(x_test))

    # Get the image and true label
    image = x_test[image_index:image_index+1]
    true_label = y_test[image_index]

    # Make prediction
    prediction_proba = model.predict(image, verbose=0)
    predicted_class = np.argmax(prediction_proba)
    confidence = np.max(prediction_proba)

    # Display results
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

    # Show image
    ax1.imshow(image[0].squeeze(), cmap='gray')
    ax1.set_title(f'True Label: {true_label}\nPredicted: {predicted_class}\nConfidence: {confidence:.3f}')
    ax1.axis('off')

    # Show probability distribution
    ax2.bar(range(10), prediction_proba[0], color='skyblue', alpha=0.7)
    ax2.bar(predicted_class, prediction_proba[0][predicted_class], color='red', alpha=0.8)
    ax2.set_title('Prediction Probabilities')
    ax2.set_xlabel('Digit')
    ax2.set_ylabel('Probability')
    ax2.set_xticks(range(10))

    # Add probability labels
    for i, prob in enumerate(prediction_proba[0]):
        ax2.text(i, prob + 0.01, f'{prob:.3f}', ha='center', va='bottom', fontsize=8)

    plt.tight_layout()
    plt.show()

    is_correct = predicted_class == true_label
    print(f"Prediction: {'✅ CORRECT' if is_correct else '❌ INCORRECT'}")

    return predicted_class, confidence, is_correct

# Test with random examples
print("Testing with random examples:")
for i in range(3):
    print(f"\n--- Example {i+1} ---")
    predict_digit(model)

# ============================================================================
# SECTION 10: Model Summary and Statistics
# ============================================================================

def print_final_summary(model, results, history):
    """Print comprehensive model summary"""

    print("="*60)
    print("MNIST CNN MODEL - FINAL SUMMARY")
    print("="*60)
    print(f"Author: sntsemilio")
    print(f"Date: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
    print(f"Environment: Google Colab")
    print()

    # Model architecture info
    total_params = model.count_params()
    trainable_params = sum([tf.keras.backend.count_params(w) for w in model.trainable_weights])

    print("MODEL ARCHITECTURE:")
    print(f"  - Type: Convolutional Neural Network (CNN)")
    print(f"  - Total Parameters: {total_params:,}")
    print(f"  - Trainable Parameters: {trainable_params:,}")
    print(f"  - Layers: {len(model.layers)}")
    print()

    # Training info
    epochs_trained = len(history.history['accuracy'])
    best_val_acc = max(history.history['val_accuracy'])
    best_epoch = np.argmax(history.history['val_accuracy']) + 1

    print("TRAINING RESULTS:")
    print(f"  - Epochs Trained: {epochs_trained}")
    print(f"  - Best Validation Accuracy: {best_val_acc:.4f} (Epoch {best_epoch})")
    print(f"  - Final Training Accuracy: {history.history['accuracy'][-1]:.4f}")
    print(f"  - Final Validation Accuracy: {history.history['val_accuracy'][-1]:.4f}")
    print()

    # Test results
    print("TEST SET PERFORMANCE:")
    print(f"  - Test Accuracy: {results['test_accuracy']:.4f}")
    print(f"  - Test Loss: {results['test_loss']:.4f}")
    print(f"  - Error Rate: {(1 - results['test_accuracy']):.4f}")
    print()

    # Per-class performance
    print("PER-CLASS ACCURACY:")
    for digit, acc in enumerate(results['class_accuracy']):
        print(f"  - Digit {digit}: {acc:.4f}")
    print()

    print("MODEL SAVED AS: 'best_mnist_model.h5'")
    print("="*60)

# Print final summary
print_final_summary(model, results, history)

# ============================================================================
# SECTION 11: Save Model and Results
# ============================================================================

# Save the final model
model.save('final_mnist_model.h5')
print("✅ Model saved as 'final_mnist_model.h5'")

# Save training history
import pickle
with open('training_history.pkl', 'wb') as f:
    pickle.dump(history.history, f)
print("✅ Training history saved as 'training_history.pkl'")

# Save results summary
summary_data = {
    'test_accuracy': results['test_accuracy'],
    'test_loss': results['test_loss'],
    'class_accuracy': results['class_accuracy'],
    'total_parameters': model.count_params(),
    'epochs_trained': len(history.history['accuracy']),
    'best_val_accuracy': max(history.history['val_accuracy']),
    'training_date': datetime.now().strftime('%Y-%m-%d %H:%M:%S')
}

with open('model_summary.pkl', 'wb') as f:
    pickle.dump(summary_data, f)
print("✅ Model summary saved as 'model_summary.pkl'")

print("\n🎉 MNIST Classification Project Complete!")
print("All files saved and ready for deployment.")
```
